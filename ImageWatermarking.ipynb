```python
red_patch = mpatches.Patch(color='red', label='Red channel')
green_patch = mpatches.Patch(color='green', label='Green channel')
blue_patch = mpatches.Patch(color='blue', label='Blue channel')

original_patch = mpatches.Patch(color='black', label='Original singular value')
watermarked_patch = mpatches.Patch(color='cyan', label='Watermarked singular value')
normalized_patch = mpatches.Patch(color='magenta', label='Normalized singular value')
```

```python
# Embedding 0
watermarked_image = watermarkImage(image,left,right,0)
```

```python
try:
    for c in range(watermarked_image.shape[2]):
        _,s_original,_ = np.linalg.svd(image[:,:,c],full_matrices=True) 
        _,s_float,_ = np.linalg.svd(watermarked_image[:,:,c],full_matrices=True)
        plt.figure()
        plt.ylabel('Singular value')
        plt.xlabel('Singular value postion')
        plt.legend(handles=[original_patch,watermarked_patch])
        plt.plot(s_original[left-10:right+10],'k*:')
        plt.plot(s_float[left-10:right+10],'co:')
    
except:
    _,s_original,_ = np.linalg.svd(image[:,:],full_matrices=True) 
    _,s_float,_ = np.linalg.svd(watermarked_image[:,:],full_matrices=True)
    plt.figure()
    plt.ylabel('Singular value')
    plt.xlabel('Singular value postion')
    plt.legend(handles=[original_patch,watermarked_patch])
    plt.plot(s_original[left-10:right+10],'k*:')
    plt.plot(s_float[left-10:right+10],'co:')
```
